```python
!pip install dash plotly pandas

import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import pandas as pd
import plotly.express as px

# Carregar o arquivo CSV
df2 = pd.read_csv('/content/df2.csv')
df2
```

```python
import pandas as pd
import plotly.graph_objs as go
from plotly.subplots import make_subplots

```

```python

fig = make_subplots(rows=1, cols=3, subplot_titles=("Nome Transação", "Status Proposta", "Faixa Etária"))

# Adicionando gráfico de barras para Nome Transação
fig.add_trace(go.Bar(x=df2['nome_transacao'].value_counts().index,
                     y=df2['nome_transacao'].value_counts().values,
                     name='Nome Transação'),
              row=1, col=1)

# Adicionando gráfico de barras para Status Proposta
fig.add_trace(go.Bar(x=df2['status_proposta'].value_counts().index,
                     y=df2['status_proposta'].value_counts().values,
                     name='Status Proposta'),
              row=1, col=2)

# Adicionando gráfico de barras para Faixa Etária
fig.add_trace(go.Bar(x=df2['faixa_etaria'].value_counts().index,
                     y=df2['faixa_etaria'].value_counts().values,
                     name='Faixa Etária'),
              row=1, col=3)

# Atualizando layout
fig.update_layout(height=400, width=1000, title_text="Dashboard")

# Exibindo o dashboard
fig.show()
```

```python

# Criando subplots
fig = make_subplots(rows=1, cols=3, subplot_titles=("Nome Transação", "Status Proposta", "Faixa Etária"))

# Adicionando gráfico de barras para Nome Transação
fig.add_trace(go.Bar(x=df2['nome_transacao'].value_counts().index,
                     y=df2['nome_transacao'].value_counts().values,
                     text=df2['nome_transacao'].value_counts().values,
                     name='Nome Transação'),
              row=1, col=1)

# Adicionando gráfico de barras para Status Proposta
fig.add_trace(go.Bar(x=df2['status_proposta'].value_counts().index,
                     y=df2['status_proposta'].value_counts().values,
                     text=df2['status_proposta'].value_counts().values,
                     name='Status Proposta'),
              row=1, col=2)

# Adicionando gráfico de barras para Faixa Etária
fig.add_trace(go.Bar(x=df2['faixa_etaria'].value_counts().index,
                     y=df2['faixa_etaria'].value_counts().values,
                     text=df2['faixa_etaria'].value_counts().values,
                     name='Faixa Etária'),
              row=1, col=3)

# Atualizando layout
fig.update_layout(height=400, width=1000, title_text="Dashboard")

# Exibindo o dashboard
fig.show()
```

```python

# Criando o gráfico de linhas temporal
fig.add_trace(go.Scatter(x=df2['data_transacao'],
                         y=df2['valor_transacao'],
                         mode='lines',
                         name='Valor Transação'),
              row=1, col=4)

# Atualizando layout
fig.update_xaxes(title_text="Data Transação", row=1, col=4)
fig.update_yaxes(title_text="Valor Transação", row=1, col=4)

# Atualizando layout geral
fig.update_layout(height=400, width=1200, title_text="Dashboard")

# Exibindo o dashboard
fig.show()
```
